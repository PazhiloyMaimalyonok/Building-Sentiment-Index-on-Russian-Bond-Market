# Эмиссия  ЛК РОДЕЛЕН, 001P-02

```python
import pandas as pd
import numpy as np
import pmdarima as pm
%pylab inline
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product

def invboxcox(y,lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))

from google.colab import drive
drive.mount('/content/drive')
```

```python
sheet_index = 0
data = pd.read_excel('drive/My Drive/cursa4/All.xlsx',sheet_name=sheet_index)
what = ['Цена закрытия, %',  'Доходность к погашению по цене закрытия, %','Объем сделок в бумагах' ]
print(what)
print('Что из этого хочешь? введи цифру')
want = what[int(input()) - 1]
data = data[~data['Цена закрытия, %'].isnull()][[want, 'Дата торгов']]
data.fillna(0, inplace = True)
data['Дата торгов'] = pd.to_datetime(data['Дата торгов'],yearfirst=True,format= '%Y-%m-%d')
data.set_index('Дата торгов', inplace=True)
data = data[data.index<='2020-04-01T00:00:00.000000000']
data.head()
```

```python
new_data = pd.read_csv('drive/My Drive/cursa4/rodelen_itog.csv')
#new_data['time'] = pd.to_datetime(new_data['time'],yearfirst=True,format= '%Y-%m-%d')
new_data.set_index('time' , inplace = True  )
data = pd.concat([data, new_data],  axis = 1).dropna()
assert data.shape[0] == data.shape[0]
data.head()
```

```python
#auto
import pmdarima as pm
stepwise_fit = pm.auto_arima(data[want],  
                             exogenous=data[['Роделен', 'рекламные упоминания Роделен']], trace=True,
                             error_action='warn'   ,  # don't want to know if an order does not work
                             suppress_warnings=True,  # don't want convergence warnings
                             stepwise=True) 
stepwise_fit.summary()
```

```python
plt.figure(figsize(15,8))
plt.subplot(211)
stepwise_fit.resid()[13:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(stepwise_fit.resid()[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(stepwise_fit.resid()[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(stepwise_fit.resid()[13:])[1])
```

```python
plt.figure(figsize(15,7))
data[want].plot()
plt.ylabel(want)
plt.title(want)
pylab.show()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data[want])[1])
```

```python
data['box'], lmbda = stats.boxcox(data[want])
plt.figure(figsize(15,7))
data['box'].plot()
plt.ylabel('Transformed ' + want)
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data['box'])[1])
```

```python
data['diff'] = data[want] - data[want].shift(1)
plt.figure(figsize(15,10))
data['diff'].plot() 
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data['diff'][13:])[1])
```

```python
ps = range(1,5)
d=0
qs = range(1,7)
parameters = product(ps, qs)
parameters_list = list(parameters)
len(parameters_list)
parameters = product(ps, qs)
parameters_list = list(parameters)
len(parameters_list)
#print('available columns', data.columns)
#ezz = str(input())

results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')
from tqdm import tqdm
for param in tqdm(parameters_list):
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(data[ezz], order=(param[0], d, param[1]), exog = data[[ 'Роделен','рекламные упоминания Роделен']],\
                                        initialization='approximate_diffuse').fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
warnings.filterwarnings('default')

result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
ps = range(1,5)
d=0
qs = range(1,7)
parameters = product(ps, qs)
parameters_list = list(parameters)
len(parameters_list)
parameters = product(ps, qs)
parameters_list = list(parameters)
len(parameters_list)
#print('available columns', data.columns)
#ezz = str(input())

results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')
from tqdm import tqdm
for param in tqdm(parameters_list):
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(data[ezz], order=(param[0], d, param[1]), exog = data[[ 'Роделен','рекламные упоминания Роделен']],\
                                        initialization='approximate_diffuse').fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
warnings.filterwarnings('default')

result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```
